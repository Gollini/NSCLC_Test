```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('/home/ivo.navarrete/Desktop/NSCLC/NSCLC_Test/checkpoints/ig_100/train_metrics.csv')
df_test = pd.read_csv('/home/ivo.navarrete/Desktop/NSCLC/NSCLC_Test/checkpoints/ig_100/test_metrics.csv')
```

```python
Fold0 = df_train.loc[df_train['fold'] == 0]
Fold1 = df_train.loc[df_train['fold'] == 1]
Fold2 = df_train.loc[df_train['fold'] == 2]
Fold3 = df_train.loc[df_train['fold'] == 3]
Fold4 = df_train.loc[df_train['fold'] == 4]
```

```python
x0, x1, x2, x3, x4 = list(range(len(Fold0))), list(range(len(Fold1))), list(range(len(Fold2))), list(range(len(Fold3))), list(range(len(Fold4)))
plt.plot(x0, Fold0['accuracy'], label='Fold0')
plt.plot(x1, Fold1['accuracy'], label='Fold1')
plt.plot(x2, Fold2['accuracy'], label='Fold2')
plt.plot(x3, Fold3['accuracy'], label='Fold3')
plt.plot(x4, Fold4['accuracy'], label='Fold4')
plt.legend()
plt.show()
```
