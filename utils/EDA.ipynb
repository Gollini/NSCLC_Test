```python
def create_animation(ims):
    # Create base plot to show images
    fig = plt.figure(figsize=(4, 4))
    plt.axis('off') # Remove axis lines
    im = plt.imshow(ims[0], cmap = 'gray')
    plt.close()

    def animate_func(i):
        im.set_array(ims[i])
        return im

    return animation.FuncAnimation(fig, animate_func, frames = len(ims), interval = 200)
```

```python
create_animation(np.transpose(roi, (2, 0, 1)))
```
